```python
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
bio_emo = []
for i in range(365):
    random_5 = random.randint(1,5)
    bio_emo.append(random_5)
print(bio_emo)
print(len(bio_emo))
```

```python
num_lst = []
emo_lst = []
for number, emotion in enumerate(bio_emo):
    num_lst.append(number + 1)
    emo_lst.append(emotion)
```

```python
for n,e in zip(num_lst, emo_lst):
    print(n,e)
```

```python
plt.plot(num_lst, emo_lst)
```

```python
plt.plot(num_lst[:10], emo_lst[:10])
```
